```python
import networkx as nx
from  matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# Simple Undirected Graph

```python
#create Graph Object
G=nx.Graph()

#add nodes one by one
G.add_node(1,)
G.add_node(2)
G.add_node(5)
G.add_node(7)
G.add_node(25)

#add edges
G.add_edge(1,2)
G.add_edge(1,7)
G.add_edge(5,2)
G.add_edge(2,7)
G.add_edge(7,5)
G.add_edge(1,25)
G.add_edge(25,7)
G.add_edge(5,1)

#draw the graph
nx.draw(G, with_labels=True)
plt.show()
```

# Simple Directed Graph

```python
#create Graph Object
G=nx.DiGraph()

#add nodes one by one
G.add_node(1,)
G.add_node(2)
G.add_node(5)
G.add_node(7)
G.add_node(25)

#add edges
G.add_edge(1,2)
G.add_edge(1,7)
G.add_edge(5,2)
G.add_edge(2,7)
G.add_edge(7,5)
G.add_edge(1,25)
G.add_edge(25,7)
G.add_edge(5,1)

#draw the graph
nx.draw(G, with_labels=True)
plt.show()
```

# Adding Node Labels, coloring, etc..

```python
#create Graph Object
G=nx.Graph()

#add nodes one by one
G.add_node(1,)
G.add_node(2)
G.add_node(5)
G.add_node(7)
G.add_node(25)

#add edges
G.add_edge(1,2)
G.add_edge(1,7)
G.add_edge(5,2)
G.add_edge(2,7)
G.add_edge(7,5)
G.add_edge(1,25)
G.add_edge(25,7)
G.add_edge(5,1)


#create string labels
labels={}
labels[1]=r'$Hello$'
labels[2]=r'$\Sigma$'
labels[5]=r'$\gamma$'
labels[7]=r'$\beta$'
labels[25]=r'$\alpha$'


nx.draw(G, labels=labels,with_labels=True, font_size=24,node_size=600, node_color="yellow")
plt.show()
```

# More complex Graphs

```python
G=nx.cubical_graph()
pos=nx.spring_layout(G) # positions for all nodes

# nodes
nx.draw_networkx_nodes(G,pos,
                       nodelist=[0,1,2,3],
                       node_color='r',
                       node_size=500,
                   alpha=0.8)
nx.draw_networkx_nodes(G,pos,
                       nodelist=[4,5,6,7],
                       node_color='b',
                       node_size=500,
                   alpha=0.8)

# edges
nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.5)
nx.draw_networkx_edges(G,pos,
                       edgelist=[(0,1),(1,2),(2,3),(3,0)],
                       width=8,alpha=0.5,edge_color='r')
nx.draw_networkx_edges(G,pos,
                       edgelist=[(4,5),(5,6),(6,7),(7,4)],
                       width=8,alpha=0.5,edge_color='b')


# some math labels
labels={}
labels[0]=r'$a$'
labels[1]=r'$b$'
labels[2]=r'$c$'
labels[3]=r'$d$'
labels[4]=r'$\alpha$'
labels[5]=r'$\beta$'
labels[6]=r'$\gamma$'
labels[7]=r'$\delta$'
nx.draw_networkx_labels(G,pos,labels,font_size=16)

plt.axis('off')
plt.savefig("labels_and_colors.png") # save as png
plt.show() # display
```
